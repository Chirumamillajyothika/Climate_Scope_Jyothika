```python
import pandas as pd

# Load the dataset
file_path = 'Global_Weather_Repository.csv'
df = pd.read_csv(file_path)
print("Dataset loaded successfully.")
```

```python
# Explore dataset using head(), info(), shape, describe()
print("First 5 rows of the dataset:")
print(df.head())
print("\nDataset Info:")
print(df.info())
print("\nDataset Shape (rows, columns):", df.shape)
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Identify missing values and duplicates
print("\nMissing Values:")
print(df.isnull().sum())
print("\nNumber of Duplicate Rows:", df.duplicated().sum())
```

```python
# Perform Univariate Analysis (distribution of each column)
print("\nUnivariate Analysis (sample distribution):")
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    print(f"\nDistribution of {column}:")
    print(df[column].value_counts().head())
```

```python
# --------------------------------------------------------------
# BIVARIATE ANALYSIS  SELECTED PAIRS (FAST & BEAUTIFUL GRID)
# --------------------------------------------------------------

import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import numpy as np

print("\n" + "="*60)
print("BIVARIATE ANALYSIS: Key Numeric Pairs")
print("="*60)

# ---------------------------------------------------------
# 1. Choose the columns you actually want to compare
# ---------------------------------------------------------
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Keep only columns that have at least 2 unique values
numeric_cols = [c for c in numeric_cols if df[c].nunique() > 1]

# ---------------------------------------------------------
# 2. Define the pairs you care about (you can add/remove)
# ---------------------------------------------------------
pairs = [
    ('temperature_celsius', 'humidity'),
    ('temperature_celsius', 'pressure_mb'),
    ('temperature_celsius', 'wind_kph'),
    ('wind_kph',           'gust_kph'),
    ('latitude',           'temperature_celsius'),
    ('longitude',          'temperature_celsius')
]

# Remove pairs where a column is missing
pairs = [(x, y) for x, y in pairs if x in numeric_cols and y in numeric_cols]

# ---------------------------------------------------------
# 3. Build a nice grid (max 33 = 9 plots)
# ---------------------------------------------------------
n_pairs = len(pairs)
cols = min(3, n_pairs)
rows = (n_pairs + cols - 1) // cols

plt.figure(figsize=(5 * cols, 4 * rows))
for idx, (x, y) in enumerate(pairs, 1):
    plt.subplot(rows, cols, idx)
    
    # Drop NaN for this pair only
    data = df[[x, y]].dropna()
    
    if len(data) == 0:
        plt.text(0.5, 0.5, 'No data', ha='center', va='center',
                 transform=plt.gca().transAxes, fontsize=12, color='gray')
    else:
        # Scatter
        sns.scatterplot(data=data, x=x, y=y, alpha=0.6, edgecolor=None, color='#2c7bb6')
        # LOWESS trend line
        sns.regplot(data=data, x=x, y=y, scatter=False,
                    lowess=True, color='red', line_kws={'lw':2})
        # Correlation
        r, _ = pearsonr(data[x], data[y])
        plt.title(f'{x} vs {y}\n$r = {r:.3f}$', fontsize=11, pad=8)

    plt.xlabel(x)
    plt.ylabel(y)

plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 4. Full correlation matrix (table + heatmap)
# ---------------------------------------------------------
print("\nPearson Correlation Matrix (all numeric columns):")
corr = df[numeric_cols].corr().round(3)
display(corr)

# Heatmap (lower triangle only)
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(10, 8))
sns.heatmap(corr, mask=mask, annot=True, fmt=".2f", cmap='coolwarm',
            center=0, linewidths=0.7, cbar_kws={"shrink": .8})
plt.title('Correlation Heatmap', fontsize=14, pad=15)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Cell: Correlation Heatmap 
print("Generating Correlation Heatmap...")

# Select only numeric columns that exist
final_numeric_cols = [col for col in numeric_cols if col in df.columns]
corr_matrix = df[final_numeric_cols].corr(method='pearson')

# Create mask for lower triangle only (cleaner look)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(16, 12))
sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,           # Show correlation values
    fmt=".2f",            # 2 decimal places
    cmap='coolwarm',      # Red = positive, Blue = negative
    center=0,             # Center color at 0
    linewidths=0.7,       # Grid lines
    linecolor='white',
    cbar_kws={"shrink": 0.8, "label": "Pearson Correlation"},
    square=True,          # Square cells
    annot_kws={"size": 10, "weight": "bold"}  # Bold numbers
)
plt.title('Correlation Heatmap of Weather Variables', fontsize=18, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print("Heatmap generated!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Detect outliers and anomalies (using IQR method for numerical columns)
print("\n=== Detecting Outliers and Anomalies ===")
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Print outlier counts using IQR method
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    print(f"\nOutliers in {column}: {len(outliers)}")

# Visualize outliers using box plots
plt.figure(figsize=(15, 5))
sns.boxplot(data=df[numerical_columns])
plt.title('Box Plot for Outlier Detection')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Remove duplicate rows
df_cleaned = df.drop_duplicates()
print("\nDuplicates removed. New shape:", df_cleaned.shape)
```

```python
# Handle missing values (example: fill with median)
for column in df_cleaned.select_dtypes(include=['int64', 'float64']).columns:
    df_cleaned[column] = df_cleaned[column].fillna(df_cleaned[column].median())
print("\nMissing values handled. Missing values count:")
print(df_cleaned.isnull().sum())
```

```python
# Convert data types (example: convert 'date' to datetime if present)
if 'date' in df_cleaned.columns:
    df_cleaned['date'] = pd.to_datetime(df_cleaned['date'])
print("\nData types converted. Updated data types:")
print(df_cleaned.dtypes)
```

```python
# Save the final cleaned dataset
df_cleaned.to_csv('Global_Weather_Repository_cleaned.csv', index=False)
print("\nCleaned dataset saved as 'Global_Weather_Repository_cleaned.csv'")
```
